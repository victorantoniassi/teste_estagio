**Média e mediana de Activites por class, grade e level**

```python
def media(col1, col2, labels):
    for label in labels:
        print(label)
        print(col1.loc[ col2 == label].mean())
        print('')
```

```python
media(df['Activities'], df['Class'], classes)
```

```python
media(df['Activities'], df['Grade'], grades)
```

```python
media(df['Activities'], df['level'], levels)
```

```python
def mediana(col1, col2, labels):
    for label in labels:
        print(label)
        print(col1.loc[ col2 == label].median())
        print('')
```

```python
mediana(df['Activities'], df['Class'], classes)
```

```python
mediana(df['Activities'], df['Grade'], grades)
```

```python
mediana(df['Activities'], df['level'], levels)
```

**Correlação entre feedbacks e speaking tasks**

```python
df2 = df[['Feedbacks','Speaking tasks']]
```

```python
df2.corr()
```

```python
ax = sns.scatterplot(x='Feedbacks', y='Speaking tasks', data=df2)
```

* Existe uma forte correlação positiva entre as duas variáveis (coeficiente de correlação r = 0.931199)
* Pensando de maneira lógica, faz sentido essa correlação entre a quantidade de tasks feitas e a quantidade de feedbacks recebidos, pode-se inferir que um usuário que faz muitas tasks, recebe muitos feedbacks.
* A alta concentração de pontos entre 0 e 50 no eixo de y, pode indicar que muitos usuários estão no começo de seus estudos e/ou não estão dedicando um tempo hábil na plataforma, o que acaba também afetando o eixo y, é necessário estudos mais detalhados para se inferir algo.
* Para se inferir com algum tipo de segurança uma relação de causalidade entre essas duas variáveis, é necessário mais estudos, principalmente com o uso de metodologias científicas.
